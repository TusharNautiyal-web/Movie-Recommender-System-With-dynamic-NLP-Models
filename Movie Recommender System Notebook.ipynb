```python
import ast 
ast.literal_eval(df['genres'][0])
```

```python
def process(element):
    names = []
    for i in ast.literal_eval(element):
        names.append(i['name'])
    return names
```

```python
df['genres'] = df['genres'].apply(process)
df['keywords'] = df['keywords'].apply(process)
```

```python
df
```

```python
df['cast'][0]
```

```python
# As i don't want all names we can break it after 4
def converter(element):
    new_list = []
    counter= 0
    for i in ast.literal_eval(element):
        if counter==4:
            break
        new_list.append(i['name'])
        counter+=1
    return new_list
```

```python
df['cast']  = df['cast'].apply(converter)
```

```python
# Lets get value for director as director are very important for movies and a good director
# Gives a higher chance of Good movie or similar movies.
ast.literal_eval(df['crew'][0])
```

```python
def converter(element):
    new_list = []
    for i in ast.literal_eval(element):
        if i['job'] == 'Director':
            new_list.append(i['name'])
    return new_list
df['crew']  = df['crew'].apply(converter)
```

# As for all nlp projects we should remove stop words

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stopwords(obj):
    final_sentence = []
    tokens = word_tokenize(obj)
    for each_word in tokens:
        if each_word not in stopwords.words('english'):
            final_sentence.append(each_word)
    return " ".join(final_sentence)
```

```python
final_df['tags'] = final_df['tags'].apply(remove_stopwords)
```

```python
final_df['tags'].head()[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    result = []
    for i in text.split():
        result.append(ps.stem(i))
    return " ".join(result)
```

```python
final_df['tags'] = final_df['tags'].apply(stemming)
```
