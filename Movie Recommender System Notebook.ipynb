```python
# As i don't want all names we can break it after 4
def converter(element):
    new_list = []
    counter= 0
    for i in ast.literal_eval(element):
        if counter==4:
            break
        new_list.append(i['name'])
        counter+=1
    return new_list
```

```python
df['cast']  = df['cast'].apply(converter)
```

# As for all nlp projects we should remove stop words

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stopwords(obj):
    final_sentence = []
    tokens = word_tokenize(obj)
    for each_word in tokens:
        if each_word not in stopwords.words('english'):
            final_sentence.append(each_word)
    return " ".join(final_sentence)
```

```python
final_df['tags'] = final_df['tags'].apply(remove_stopwords)
```

```python
final_df['tags'].head()[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    result = []
    for i in text.split():
        result.append(ps.stem(i))
    return " ".join(result)
```

```python
word_embeddings = []
```

```python
def vectors():
    for line in final_df['tags']:
        avgword2vec = None
        count = 0
        for word in line.split():
            if word in model.wv:
                count += 1
                if avgword2vec is None:
                    avgword2vec = model.wv.get_vector(word)
                else:
                    avgword2vec = avgword2vec + model.wv.get_vector(word)
                
        if avgword2vec is not None:
            avgword2vec = avgword2vec / count
            word_embeddings.append(avgword2vec)
```

```python
vectors()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_with_word_2_vec = cosine_similarity(word_embeddings)
```

```python
def recommendations_with_word2vec(movie):
    # Calling the function vectors
    # Finding cosine similarity for the vectors
    movie = movie.lower()
    index = None
    titles = final_df['title_y']
    if movie in titles.to_list():
        index = final_df[final_df['title_y'] == movie].index[0]
    else:
        index = find_closest_1(movie)
    
    distances = sorted(list(enumerate(similarity_with_word_2_vec[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(final_df.iloc[i[0]].title_y)
    
```

# Now let's try Combination of Two TF-IDF + Word2Vec

```python
# So lets start with building tifidf model again which is not recommended but 
# just for the sake of easy understanding iam doing it
from sklearn.feature_extraction.text import TfidfVectorizer
stopset = set(stopwords.words('english'))
tf_idf = TfidfVectorizer(use_idf = True, lowercase = True, strip_accents='ascii',analyzer='word', ngram_range=(1, 3), min_df = 5,stop_words=stopset)
tf_idf.fit_transform(final_df['tags'])

# Getting words from TF-IDF.
tfidf_list = dict(zip(tf_idf.get_feature_names(), list(tf_idf.idf_)))
tfidf_feature = tf_idf.get_feature_names()
```

```python
# Building TF-IDF Word2Vec and then
# Storing the TFIDF Word2Vec embeddings
tfidf_and_word2vec_vectors = []
lines = 0
for line in final_df['tags']: 
    sent_vec = np.zeros(100) 
    weight_sum =0; 
    for word in line.split(): 
        if word in model.wv and word in tfidf_feature:
            vec = model.wv.get_vector(word)
            tf_idf = tfidf_list[word] * (line.count(word) / len(line))
            sent_vec += (vec * tf_idf)
            weight_sum += tf_idf
    if weight_sum != 0:
        sent_vec /= weight_sum
    tfidf_and_word2vec_vectors.append(sent_vec)
    lines+=1
```

```python
tfidf_and_word2vec_vectors
```

```python
lines
```

```python
similarity_with_tf_idf_word_2_vec  = cosine_similarity(tfidf_and_word2vec_vectors)
```

```python
def recommendations_with_tfidf_and_word2vec(movie):
    # Calling the function vectors
    # Finding cosine similarity for the vectors
    movie = movie.lower()
    index = None
    titles = final_df['title_y']
    if movie in titles.to_list():
        index = final_df[final_df['title_y'] == movie].index[0]
    else:
        index = find_closest_1(movie)
    
    distances = sorted(list(enumerate(similarity_with_tf_idf_word_2_vec[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(final_df.iloc[i[0]].title_y)
```

# Creating Sentiment Analysis function and Saving it to use.
